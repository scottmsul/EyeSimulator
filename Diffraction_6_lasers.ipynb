# Introduction to Lasers

Some of the ideas I have for measuring myopia involve lasers, so I figured it's probably worth understanding them better. I was also curious to understand them from a diffraction standpoint. I was familiar with the fact that when you hold a hair in front of a laser it creates a diffraction pattern on the wall, and I wanted to better understand the math involved in this case. As it turns out lasers can be fully solved analytically which I found interesting, however it is a little bit of work.

## Rayleigh-Sommerfeld Diffraction Integral with a Gaussian Center

As usual we start with the Rayleigh-Sommerfeld diffraction integral for propagation of light in free space, ignoring the $e^{ikz}$ term:

$$\psi(x,y,z) = - \frac{ik}{2\pi} \frac{1}{z} \exp{ \left[ \frac{ik}{2z}(x^2+y^2) \right] } \iint_{-\infty}^{\infty} \psi(x',y') \exp{ \left[ \frac{ik}{2z}(x'^2 + y'^2) \right] }  \exp{ \left[ \frac{-ik}{z}(xx'+yy') \right] }dx' dy'$$

To derive the TEM00 mode of an ideal laser, we assume that in the XY plane the amplitude follows a Gaussian dropoff centered at the origin with the same phase everywhere in the plane, as done in [Principles of Lasers](https://www.chemistry.uoc.gr/lapkin/OrazioSvelto_PrinciplesOfLasers_5thEd.pdf). Note the laser book (and Wikipedia) use the opposite convention as Goodman with $-ik$ instead of $ik$ in the spatial solution and $i \omega t$ instead of $-i \omega t$ in the time solution. I will stick with the Goodman convention to be consistent with the other notebooks.

$$\psi(x',y') = \psi_0 \exp{ \left[ - \left( \frac{x'^2 + y'^2}{w_0^2} \right) \right] }$$

Note that $w_0$ is a purely real number. Part of the motivation for having a constant phase in the $z=0$ plane is that a laser is generated by exciting light in a symmetrical cavity, so we'd expect the center to be flat in terms of phase. 

Since both the integral and amplitude share the same $(x'^2+y'^2)$ term, we can simplify the integral by defining an imaginary parameter $q$ using $1/w_0^2 = -i k / 2 q$:

$$\psi(x',y') = \psi_0 \exp{ \left[ \frac{i k}{2 q} \left( x'^2 + y'^2 \right)\right] }$$

Now the $1/q$ term can be combined with the $1/z$ term inside the exponent, since they both multiply the same factor of $\frac{i k}{2 q} \left( x'^2 + y'^2 \right)$:

$$\psi(x,y,z) = - \psi_0 \frac{ik}{2\pi} \frac{1}{z} \exp{ \left[ \frac{ik}{2z}(x^2+y^2) \right] } \iint_{-\infty}^{\infty} \exp{ \left[ \frac{i k}{2} \left( \frac{1}{q} + \frac{1}{z} \right) \left( x'^2 + y'^2 \right)\right] } \exp{ \left[ \frac{-ik}{z}(xx'+yy') \right] }dx' dy'$$

### Gaussian Integrals

Note that the $x'$ and $y'$ integrals can be fully separated:

$$\psi(x,y,z) = - \psi_0 \frac{ik}{2\pi} \frac{1}{z} \exp{ \left[ \frac{ik}{2z}(x^2+y^2) \right] } \text{Gauss}(x) \text{Gauss}(y)$$

Where:

$$\text{Gauss}(u) = \int_{-\infty}^{\infty} \exp{ \left[ \frac{i k}{2} \left( \frac{1}{q} + \frac{1}{z} \right) u'^2 \right] } \exp{ \left[ \frac{-ik}{z}uu' \right] } du'$$

This Gaussian Integral has the following [closed form solution](https://www.wolframalpha.com/input?i=integral%28e%5E%28-%28ax%5E2%2Bbx%29%29%29+from+x%3D-infinity+to+infinity):

$$\int_{-\infty}^{\infty} \exp{ \left[ - \left( ax^2 + bx \right) \right] } dx = \sqrt{\frac{\pi}{a}} \exp{ \left[ \frac{b^2}{4 a} \right] }$$

Where the $a$ and $b$ terms are given by:

$$\begin{align}
a &= \frac{-ik}{2} \left( \frac{1}{q} + \frac{1}{z} \right) \\
b &= \frac{ikx}{z}, \frac{iky}{z}
\end{align}$$

The amplitudes from the $x$ and $y$ integrals have the same $a$ term, so they multiply to give the following:

$$\sqrt{\frac{\pi}{a_{x}}} \cdot \sqrt{\frac{\pi}{a_{y}}} = \frac{\pi}{a} = -\frac{2 \pi}{ik} \left( \frac{1}{q} + \frac{1}{z} \right)^{-1}$$

The exponential $b^2/4a$ terms end up being much more complicated and interesting, and a big part of this notebook is dedicated to solving them. First note that they add since they are inside multiplied exponentials:

$$\frac{b^2}{4a_{x}}+\frac{b^2}{4a_{y}} = -\left( \frac{ik}{z} \right)^2 \left( x^2 + y^2 \right) \left( \frac{1}{2 i k} \right)\left( \frac{1}{q} + \frac{1}{z} \right)^{-1}$$

This can be cleaned up by cancelling one $ik$ and moving one $z^{-1}$ into the $\left( \frac{1}{q} + \frac{1}{z} \right)^{-1}$ term:

$$\frac{b^2}{4a_{x}}+\frac{b^2}{4a_{y}} = - \frac{ik}{2z} \left(x^2+y^2\right) \frac{1}{1+(z/q)}$$

Note that since $q$ is a pure imaginary number, we can re-write the fraction as separate real and imaginary parts:

$$\frac{1}{1+(z/q)} = \frac{1}{1+(z/q)} \cdot \frac{1 - (z/q)}{1-(z/q)} = \frac{1}{1-(z/q)^2} - \frac{(z/q)}{1-(z/q)^2}$$

The term inside the exponential becomes:

$$\frac{b^2}{4a_{x}}+\frac{b^2}{4a_{y}} = - \frac{ik}{2z} \left(x^2+y^2\right) \left[ \frac{1}{1-(z/q)^2} - \frac{(z/q)}{1-(z/q)^2} \right]$$

We will look at each term separately below. Note that since $q$ is imaginary and the exponential already has an $i$ in it, the first term is imaginary while the second term is real. 

### Real Exponential Term

Let's focus on the second (real) term first. Note there's two minus signs that cancel:

$$\text{real exp} = \frac{ik}{2z} \left(x^2+y^2\right) \frac{(z/q)}{1-(z/q)^2}$$

Recall from earlier that $1/q = - 2 / i k w_0^2$. To simplify this we can define a parameter $z_r = |q| = k w_0^2 / 2$ called the Rayleigh Distance. This parameter has interesting theoretical properties, as it's the distance at which the behavior of the laser starts to change qualitatively from a straight beam to a slowly-expanding cone. As we'll see below for a typical laser pointer this distance is actually on human scales (i.e. around a few meters). We can substitute $1/q = - 2 / i k w_0^2$ in the numerator as most of the terms cancel, and substitute $q = - i z_r$ in the denominator:

$$\text{real exp} = - \frac{1}{w_0^2} \left( \frac{x^2+y^2}{1+(z/z_r)^2} \right)$$

This can be simplified by defining a function called the "beam width":

$$w(z) = w_0 \sqrt{ 1 + (z/z_r)^2 }$$

This starts out as $w_0$ in the center plane, but grows hyperbolically with $z/z_r$ as we travel forward in the $z$ direction. With this new term the exponent becomes:

$$\text{real exp} = - \frac{ x^2 + y^2 }{w(z)^2}$$

Note that $w(z)$ is analogous to $\sigma$ in a Normal Distribution, and is also the distance from the $z$-axis at which the amplitude falls to $1/e^2$ of the amplitude at its center. For this reason, $D=2w(z)$ is referred to as the [$1/e^2$ width](https://en.wikipedia.org/wiki/Beam_diameter#1/e2_width).

### Imaginary Exponential Term

The second term becomes:

$$\text{imag exp} = - \frac{ik}{2z} \left(x^2+y^2\right) \frac{1}{1+(z/z_r)^2}$$

We can combine this term with the original term outside the integral to get the following:

$$\text{imag exp} = ik \left( \frac{x^2+y^2}{2z} \right) \left[ 1 - \frac{1}{1+(z/z_r)^2} \right]$$

We can simplify the term in square brackets:

$$1 - \frac{1}{1+(z/z_r)^2} = \frac{(z/z_r)^2}{1+(z/z_r)^2} = \frac{1}{(z_r/z)^2+1}$$

The exponent becomes:

$$\text{imag exp} = ik \left[ \frac{x^2+y^2}{2 z \left( 1 + (z_r/z)^2 \right)} \right]$$

In the laser literature this denominator is re-written as follows. Suppose we have a spherical wave emanating from the origin with radius $R$. Then along the $z$ axis at a distance $z=R$, under the Paraxial Approximation the phase shift at $(x,y)$ is given by the following. Note this is derived in previous notebooks, i.e. [Diffraction Part 3](./Diffraction_3_Goodman_Ch4_Fresnel_Fraunhofer.ipynb).

$$\text{spherical wave phase shift} = ik \left[ \frac{x^2+y^2}{2z} \right] = ik \left[ \frac{x^2+y^2}{2R} \right]$$

However in our case instead of $z$ we have $z \left[ 1 + (z_r/z)^2 \right]$ instead. This means to an observer of a laser at a distance $z$ the phase will appear curved as if it was emanating from the following radius instead:

$$R(z) = z \left[ 1 + (z_r/z)^2 \right]$$

This term is called the "Radius of Curvature." Under this definition the exponential phase term becomes:

$$\text{imag exp} = ik \left( \frac{x^2+y^2}{2 R(z)} \right)$$

### Remaining Amplitude Terms

Bringing our attention to the remaining amplitude terms, the solution is given by the following:

$$\psi = - \psi_0 \frac{ik}{2 \pi} \frac{1}{z} \cdot \left( \frac{\pi}{a} \right) \cdot (\text{Real and Imag Exp Terms})$$

Where the $\pi/a$ term was from the Gaussian Integral:

$$\frac{\pi}{a} = \frac{2 \pi}{i k} \left( \frac{1}{q} + \frac{1}{z} \right)^{-1}$$

The $2 \pi / i k$ terms cancel, and we can fold the $z^{-1}$ into the term in parantheses, giving us the following:

$$\psi = - \psi_0 \frac{1}{1 + (z/q)} \cdot (\text{Real and Imag Exp Terms})$$

Using the same algebra as before we can split the term in parantheses into its real and imaginary parts:

$$\frac{1}{1 + (z/q)} = \frac{1}{1 + (z/q)} \cdot \frac{1-(z/q)}{1-(z/q)} = \frac{1}{1+(z/z_r)^2} + \frac{(z/z_r)i}{1+(z/z_r)^2}$$

Since this is an amplitude and not an exponent, it's more useful to write this complex number using $a+bi \to A e^{i \phi}$, where:

$$\begin{align}
A &= \sqrt{a^2 + b^2} \\
\phi &= \text{atan} (b/a)
\end{align}$$

The term becomes:

$$A = \sqrt{ \frac{ 1+(z/z_r)^2 }{ (1+(z/z_r)^2)^2 } } = \left[ 1 + (z/z_r)^2 \right]^{-1/2} = \frac{w_0}{w(z)} $$

$$\phi(z) = \text{atan} (z/z_r)$$

This extra phase shift is called the "Guoy Phase" on Wikipedia.

## Solution

Using everything built up we have:

$$\psi(x,y,z) = \psi_0 \frac{w_0}{w(z)} \exp{ \left[ - \frac{ x^2+y^2 }{w(z)^2} + i k \frac{x^2+y^2}{2 R(z)} + i \phi(z) \right] }$$

Where:

$$\begin{align}
w(z) &= w_0 \sqrt{ 1 + (z/z_r)^2 } \\
R(z) &= z \left[ 1 + (z_r/z)^2 \right] \\
\phi(z) &= \text{atan} (z/z_r) \\
z_r &= k w_0^2 / 2
\end{align}$$

Note that for a typical handheld laser pointer we have the following:

$$\begin{align}
w_0 &\approx 1mm \approx 10^{-3}m \\
z_r &= \frac{k w_0^2}{2} \approx \frac{10^7 \cdot (10^{-3})^2}{2}m \approx 5m
\end{align}$$

Because these are typical human scales I don't think we can safely assume either $z>>z_r$ or $z<<z_r$.

## Example Wavefront

As it turns out the TEM00 Gaussian Mode of light isn't just for lasers, but can be generated on many different wavelengths of the electromagnetic spectrum. The Wikipedia page for [Gaussian Beams](https://en.wikipedia.org/wiki/Gaussian_beam) shows a plot of the amplitude for a radio wave with $\lambda = 30cm$ and $w_0 = 40cm$ which is reproduced below:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors as c

wavelength = .03
w0 = wavelength * (4/3)
k = 2*np.pi / wavelength
N = 1000
Lx = wavelength*20
Lz = wavelength*40

dx, dz = np.mgrid[-N/2:N/2+1:1, -N/2:N/2+1:1]
dx *= Lx / N
dz *= Lz / N
zr = k * w0**2 / 2
w = w0 * np.sqrt(1 + (dz/zr)**2)
R = np.zeros(dz.shape)
R[dz!=0] = dz[dz!=0] * (1 + (zr/dz[dz!=0])**2)
guoy = np.zeros(dz.shape)
guoy[dz!=0] = np.arctan2(dz[dz!=0],zr)
amplitude = (w0/w) * np.exp(1j*k*dz -(dx**2/w**2) + 1j * k * (dx**2/(2*R)) + 1j*guoy)
plt.pcolormesh(dz, dx, np.real(amplitude), cmap='bwr')
plt.xlabel('z')
plt.ylabel('x')
plt.show()
```
